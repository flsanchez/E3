```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from sympy import oo
from __future__ import division

sp.init_printing()
%matplotlib inline

r = sp.Symbol('r', positive=True)
k = sp.Symbol('k', positive=True)
l = sp.Symbol('l', positive=True, integer=True)
n = sp.Symbol('n', positive=True, integer=True)
z = sp.Symbol('z', positive=True, integer=True)
```

# Defino la funcion de onda radial

```python
def R_hyp(n,l,z,r):
    rho = 2 * r * z / n
    rnum = (2 * z / n)**3 * sp.factorial(n + l)
    rden = (sp.factorial(2*l+1))**2 * 2*n * sp.factorial(n-l-1)
    rnorm = sp.sqrt(rnum/rden)
    rfunc = (rho)**l * sp.exp(-rho/2) * sp.hyper((-n+l+1,), (2*l+2,), (rho))
    R = rnorm * rfunc
    return R
```

# Defino los vectores para el plot

```python
nlen = 500
xmax = 10.0
xmin = 0.0
x = np.linspace(xmin,xmax,nlen)
psi_1 = np.zeros(nlen)
psi_2 = np.zeros(nlen)
```

# Evaluo para n = 2, l = 0, Z = 3

```python
for i in range(nlen):
    psi_1[i] = R_hyp(2,0,3,x[i])
```

# Plot para n = 2, l = 0, Z = 3

```python
plt.plot(x, x*psi_1,'r')
plt.title("Radial Wavefunction $\psi_{1s}$ of Hydrogen")
plt.xlabel("r (a.u.)")
plt.ylabel("$\psi_{1s}$")
plt.show()
```

# Evaluo para n = 10, l = 3, Z = 4

```python
for i in range(nlen):
    psi_2[i] = R_hyp(10,3,4,x[i])
```

# Plot para n = 10, l = 3, Z = 4

```python
plt.plot(x, x*psi_2,'r')
plt.title("Radial Wavefunction $\psi_{3f}$ of Hydrogen")
plt.xlabel("r (a.u.)")
plt.ylabel("$\psi_{3f}$")
plt.show()
```

# Chequeo ortonormalidad

```python
sp.integrate((r*R_hyp(2,0,3,r) * r*R_hyp(10,3,4,r)),(r,0,oo)).evalf()
#sp.integrate( (r*R_hyp(1,0,1,r))**2, (r,0,oo)).evalf()
```
